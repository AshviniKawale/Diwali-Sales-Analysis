# EDA(Exploring the data and generating insights)
# Exploratory Data Analysis

# Gender


```python
# plotting a countplot for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount
# Grouping and summing the data
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Create the bar plot
plt.figure(figsize=(7, 6))
plt.bar(sales_gen['Gender'], sales_gen['Amount'], color=['blue', 'green'])
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.title('Total Amount by Gender')
plt.show()
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

# Age

```python
# Creating count plot for Age group by Gender type (Male vs Female)
ax = sns.countplot(data=df, x='Age Group', hue='Gender')

# Adding bar labels
for bars in ax.containers:
    ax.bar_label(bars)

plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Count of Age Group by Gender')
plt.show()
```

```python
# Total Amount VS Age group
# Grouping and summing the data
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age, palette='viridis')

# Adding labels on each bar

plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.title('Total Amount by Age Group')
plt.show()
```

Here, We can analyse that most of the buyers are of age group between 26-35 yrs (Youth population) female and very less buyers are from 0-17 age group (children).

# State

```python
# ceating bar plot to see total no. of orders from top 10 states
# Grouping and summing the data
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(15, 5))
ax = sns.barplot(data=sales_state, x='State', y='Orders', palette='viridis')

# Adding labels on each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

plt.xlabel('State')
plt.ylabel('Total Orders')
plt.title('Total Orders from Top 10 States')
plt.show()
```

```python
# Sales/Amount from top 10 states
# Grouping and summing the data to get the top 10 states by amount
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Setting the figure size
plt.figure(figsize=(15, 6))

# Creating the bar plot
ax = sns.barplot(data=sales_state, x='State', y='Amount', palette='viridis')

# Adding labels on each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', xytext=(0, 4), textcoords='offset points')
# Adding labels and title
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.title('Total Amount from Top 10 States')
plt.show()
```

Here, we can see that higest number of orders are from Uttar Pradesh followed by Maharashta, karnatka and lowest number of orders are from Gujrat under top 10 states.

# Marital Status

```python
# Count the occurrences of each category in 'Marital_Status'
counts = df['Marital_Status'].value_counts().sort_index()

# Set the figure size
plt.figure(figsize=(5, 5))

# Create the bar plot
bars = plt.bar(counts.index, counts.values, color=('skyblue', 'pink'))

# Adding labels and title
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title('Count of Marital Status')
plt.xticks(ticks=counts.index, labels=['Single', 'Married'])  
# Display the plot
plt.show()
```

From above graph it can be seen that single people has done lot of shpping

```python
# Sales Amount by Marital Status and Gender
# Grouping and summarizing the data
sales__marital = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set the figure size
sns.set(rc={'figure.figsize': (6, 5)})

# Create the bar plot
ax = sns.barplot(data=sales__marital, x='Marital_Status', y='Amount', hue='Gender', palette='viridis')

# Add labels on each bar
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.2f')

# Customize the plot
plt.xlabel('Marital Status')
plt.ylabel('Total Amount')
plt.title('Total Sales Amount by Marital Status and Gender')
plt.xticks(ticks=[0, 1], labels=['Single', 'Married'])  # Assuming 0=Single and 1=Married
plt.legend(title='Gender')
plt.show()
```

Here, we can see that most of the buyers are married women and single women as compared to Males, Females have high purchasing power than Males

# Occupation

```python
# count plot of Occupation
# Create the count plot
ax = sns.countplot(data=df, x='Occupation', palette='viridis')

# Add labels on each bar
for bars in ax.containers:
    ax.bar_label(bars)

# Customize the plot
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Count of Occupations')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
# Sales Amount by Occupation
# Group and summarize the data
occ_status = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set the figure size and create the bar plot
sns.set(rc={'figure.figsize': (20, 5)})
ax = sns.barplot(data=occ_status, x='Occupation', y='Amount', palette='viridis')

# Add labels on each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', xytext=(0, 4), textcoords='offset points')
    
# Customize and show the plot
plt.xlabel('Occupation')
plt.ylabel('Total Amount')
plt.title('Total Sales Amount by Occupation')
plt.xticks(rotation=45)
plt.show()
```

Here, we can see that most of the buyers are working in IT sector, Healthcare and Aviation sector where as low number of buyers are from Texlile and Agriculture.

# Product Category

```python
# count plot of Product_Category
# Create the count plot
ax = sns.countplot(data=df, x='Product_Category', palette='viridis')

# Add labels on each bar
for bars in ax.containers:
    ax.bar_label(bars)

# Customize the plot
plt.xlabel('Product_Category')
plt.ylabel('Count')
plt.title('Count of Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales Amount by Product_Category
# Group and summarize the data
prod_cat = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set the figure size and create the bar plot
sns.set(rc={'figure.figsize': (20, 5)})
ax = sns.barplot(data=prod_cat, x='Product_Category', y='Amount', palette='viridis')

# Add labels on each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', xytext=(0, 4), textcoords='offset points')
    
# Customize and show the plot
plt.xlabel('Product_Category')
plt.ylabel('Total Amount')
plt.title('Total Sales Amount by Product_Category')
plt.xticks(rotation=45)
plt.show()
```

Here, we can observe that most of the sold products are from Food, Clothing & Apparel and Electronics & Gadgets category, where as lowest sales are from veterinary and office category.
